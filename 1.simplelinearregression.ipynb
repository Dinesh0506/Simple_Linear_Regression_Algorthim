```python
#import the libary and pandas is used to numerical data fuction 
import pandas  as pd
```

```python
#usind pandas to read a data and assign to dataset varible  &print the data having dataset variablle
dataset=pd.read_csv('Salary_Data.csv')
dataset
```

```python
#split the input data and print the data 
indepentent=dataset[["YearsExperience"]]
indepentent
```

```python
#split the output data and print the data
depentent=dataset[["Salary"]]
depentent
```

```python
#split the train data and test data using function train_test_split undert sklearn function 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indepentent,depentent, test_size = 1/3, random_state = 0)
```

```python
#inthis model creation of linear model and drive the linear equation y=W*x1+b0
from sklearn.linear_model import LinearRegression 
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
 #calculate the weight of the model
weight=regressor.coef_
```

```python
weight
```

```python
  #calculate the  bias of the model
bias=regressor.intercept_
bias
```

```python
#predict the value based on evalution
y_pred=regressor.predict(X_test)
```

```python
#validating the model
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#save the model 
import pickle
filename="finalized_model.sav"
pickle.dump(regressor,open(filename, 'wb'))
#get the input from user to predict the salary #deployment phase
loaded_model = pickle.load(open('finalized_model.sav', 'rb'))
result = loaded_model.predict([[15]])
result
```
